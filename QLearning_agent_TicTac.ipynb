```python
import numpy as np
def check_over(x):
    """
    Takes in the current board state x,
    and determines if the game is won by 
    any player or if there is a draw.    
    Returns True if episode is over. 
    Returns False if episode is not over. 
    Also, returns a string '1' if player 1
    has won, a string '2' if player 2 has
    won, or a string '3' if there is a draw. 
    """
    # CHECK WHETHER PLAYER 1 WON. 
    # CHECK EACH ROW: 
    if (np.prod(x[0,:]) == 1):
        print("PLAYER 1 WON!")
        isDone = True
        player_win = '1'
        return isDone, player_win
    if (np.prod(x[1,:]) == 1):
        print("PLAYER 1 WON!")
        isDone = True
        player_win = '1'
        return isDone, player_win
    if (np.prod(x[2,:]) == 1):
        print("PLAYER 1 WON!")
        isDone = True
        player_win = '1'
        return isDone, player_win
    # CHECK EACH COLUMN:
    if (np.prod(x[:,0] == 1)):
        print("PLAYER 1 WON!")
        isDone = True
        player_win = '1'
        return isDone, player_win
    if (np.prod(x[:,1] == 1)):
        print("PLAYER 1 WON!")
        isDone = True
        player_win = '1'
        return isDone, player_win
    if (np.prod(x[:,2] == 1)):
        print("PLAYER 1 WON!")
        isDone = True
        player_win = '1'
        return isDone, player_win
    # CHECK EACH DIAGONAL
    # CHECK MAIN DIAGONAL:
    if (np.prod(np.diag(x)) == 1):
        print("PLAYER 1 WON!")
        isDone = True
        player_win = '1'
        return isDone, player_win
    # CHECK OPPOSITE DIAGONAL:
    if (np.prod(np.diag(np.rot90(x))) == 1):
        print("PLAYER 1 WON!")
        isDone = True
        player_win = '1'
        return isDone, player_win
####################################################################
    # CHECK WHETHER PLAYER 2 WON. 
    # CHECK EACH ROW: 
    if (np.sum(x[0,:]) == 6):
        print("PLAYER 2 WON!")
        isDone = True
        player_win = '2'
        return isDone, player_win
    if (np.sum(x[1,:]) == 6):
        print("PLAYER 2 WON!")
        isDone = True
        player_win = '2'
        return isDone, player_win
    if (np.sum(x[2,:]) == 6):
        print("PLAYER 2 WON!")
        isDone = True
        player_win = '2'
        return isDone, player_win
    
    # CHECK EACH COLUMN:
    if (np.sum(x[:,0] == 6)):
        print("PLAYER 2 WON!")
        isDone = True
        player_win = '2'
        return isDone, player_win
    if (np.sum(x[:,1] == 6)):
        print("PLAYER 2 WON!")  
        isDone = True
        player_win = '2'
        return isDone, player_win
    if (np.sum(x[:,2] == 6)):
        print("PLAYER 2 WON!")
        isDone = True
        player_win = '2'
        return isDone, player_win
    # CHECK EACH DIAGONAL
    # CHECK MAIN DIAGONAL:
    if (np.sum(np.diag(x)) == 6):
        #print("PLAYER 2 WON!")
        isDone = True
        player_win = '2'
        return isDone, player_win
    # CHECK OPPOSITE DIAGONAL:
    if (np.sum(np.diag(np.rot90(x))) == 6):
        #print("PLAYER 2 WON!")
        isDone = True
        player_win = '2'
        return isDone, player_win
####################################################################
    # CHECK TO SEE IF THERE IS A DRAW. 
    if np.all(x):
        #print("Draw!")
        isDone = True
        player_win = '3'
        return isDone, player_win
    
    return False, ' '
```

2-b)  Two opponents for the Q-learning agent to train against, namely, a random agent and safe agent

```python
player_markers = {1: "X", 2: "0"}
### Set o-learning data and learning parameters. # Q-matrix for player 1:

num_values = 3 
max_num_controls = 9 
Q = np.zeros((num_values, num_values, num_values,num_values, num_values, num_values, num_values, num_values, num_values, max_num_controls))
```

```python
alpha = 0.9
gamma = 0.99
epsilon = 1.0
NUM_EPISODES = 100000

# Keep track of tally for player 1 wins, player 2 wins, and draws
player1_wins = np.zeros((NUM_EPISODES,)) 
player2_wins = np.zeros((NUM_EPISODES,)) 
draws = np.zeros((NUM_EPISODES, ))
# Number of decay stages for epsilon. 
NUM_STAGES = 20

STAGE_LENGTH = NUM_EPISODES // NUM_STAGES 
# Decay factor for epsilon within each stage.
DECAY_FACTOR = 0.7

print("Shape of our Tic-Tac-Toe Q-table =", Q.shape)
```

```python
def get_control_train(x, epsilon, Controls, Q):

# print("Q.shape = ", Q.shape) 
	if np. random.random() >= epsilon:
		u = np.argmax (Q[x]) # Check to see if u is in available controls first.
		if u not in Controls:
			new_control = np.random.choice(list(Controls))
			return new_control 
		else:
			return u 
	else:
		# Choose a random control 
		u = np.random.choice(list(Controls)) 
		return u


def get_control_test(x, Controls, Q):

	u = np.argmax(Q[x])
	# Check to see if u is in available controls first... 
	if u not in Controls:
		new_control = np. random.choice(list(Controls))
		return new_control 
	else:
		return u
		

def decay_epsilon(episode,epsilon,STAGE_LENGTH, DECAY_FACTOR):
	# Decay epsilon if in a new stage: 
	if (episode % STAGE_LENGTH == 0 and episode != 0):
		epsilon *= DECAY_FACTOR 
		#print(f"epsilon = {epsilon}")
	return epsilon


def reset_controls ():
	Controls = set()
	for i in range(9):
		Controls.add(i)
	return Controls
	

def player1_batch_learn(player1_states,player1_controls, r, player1_next_states, gamma, alpha, Q):

	for i, x in enumerate(player1_states):
		u = player1_controls[i] 
		x_next = player1_next_states [i] 
		td_target = r + gamma*np. max(Q[x_next]) 
		td_error = td_target - Q[x][u] 
		Q[x][u] += alpha*td_error
	return Q 
```

```python

r = 10
for episode in range(NUM_EPISODES):
  # Hold playerl states and controls for better updating of Q-fcn: 
  player1_states = [] 
  player1_controls = [] 
  player1_next_states = []
  isDone = False
  Controls = reset_controls ()
  x = np.zeros((3,3)) 
  k = 0  
  board = np.array([['-', '-', '-'],['-', '-', '-'],['-', '-', '-']]) 
  # Check to see if we need to decay epsilon: 
  epsilon = decay_epsilon (episode, epsilon,STAGE_LENGTH, DECAY_FACTOR)
  #epsilon = 1
  
  # print (f"Episode#{episode), k={k}, \n", board)
  while (isDone != True): 
    if(k % 2 == 0):
      # Agent 1 moves first. Random agent
      epsilon = 0
      u = get_control_train(tuple(x.reshape(-1,).astype (int)),epsilon, Controls, Q)
      player1_states.append(tuple(x. reshape(-1,).astype (int)))
      player1_controls.append(u)
      Controls.remove(u) 
			# Update board state. 
      x_next = x. reshape(-1,) 
      x_next [u] = 1 
      x_next = x_next.reshape(3,3) 
      player1_next_states.append(tuple(x_next.reshape(-1,).astype(int)))
      # Update board visualization. 
      board = board.reshape(-1,) 
      board[u] = player_markers [1] 
      board = board.reshape(3,3)
      # Check if episode is over: 
      isDone, result = check_over(x) 
      if (isDone == True and result == '1'):
        player1_wins[episode] = 1 
        # Player 1 learns from his moves that lead to a win. 
        Q = player1_batch_learn(player1_states, player1_controls, r, player1_next_states, gamma, alpha, Q)
        
      elif (isDone == True and result == '3'):
        draws [episode] = 1
        # Reset the state. 
      x = x_next
      k+=1

    else:  # It is agent 2's move.
      #u = np. random.choice(list(Controls))
      epsilon = 1 # safe agent
      u = get_control_train(tuple(x.reshape(-1,).astype (int)),epsilon, Controls, Q)
      Controls.remove(u)
      x = x.reshape(-1,)
      x[u] = 2
      x = x.reshape(3,3)
      # Update board visualization.
      board = board.reshape(-1,)
      board[u] = player_markers[2]
      board = board.reshape(3,3)
      k+=1

      # Check if episode is over:
      isDone, result = check_over(x)
      if (isDone == True and result == '2'):
        player2_wins[episode] = 1
      elif (isDone == True and result == '3'):
        draws[episode] = 1 
```

```python
player1_win_total = np.cumsum(player1_wins)
player2_win_total = np.cumsum(player2_wins)
draw_total = np.cumsum (draws)

player1_win_rate = np.zeros(NUM_EPISODES,)
player2_win_rate = np.zeros(NUM_EPISODES, )
draw_rate = np.zeros(NUM_EPISODES, )

for k in range(NUM_EPISODES):
	player1_win_rate[k] = player1_win_total[k]/(k+1) 
	player2_win_rate[k] = player2_win_total[k]/(k+1) 
	draw_rate[k] = draw_total[k]/(k+1) 
```

```python
plt.figure()
plt.plot(np.arange (NUM_EPISODES), player1_win_rate, 'g-.',label='Random agent Win Rate') 
plt.plot(np.arange (NUM_EPISODES), player2_win_rate, 'r:.',label='Safe agent Win Rate')
plt.plot(np.arange (NUM_EPISODES), draw_rate, 'k-',label='Draw Rate')
plt.xlabel('Training Episodes')
plt.legend() 
```
